```python
import requests
import sys
from podcast_ad_skipper.google_cloud import *
from podcast_ad_skipper.data_preparation import *
from podcast_ad_skipper.params import *
```

```python
bq_client = auth_gc_bigquery()
```

```python
table_id = f"{GCP_PROJECT_ID}.{BQ_DATASET}.test10"
```

```python
output= get_output_query_bigquery(bq_client, table_id, limit=1, columns="*")
```

```python
processed_output = get_bq_processed_data(output)
```

```python
spectrogram_np = np.array(processed_output[0])
```

```python
try:
    # To update URL and save it as an env variable
    api_url = SERVICE_URL

    params = {"spectrogram": spectrogram_np}

    response = requests.get(api_url, params=params)

    # Check if the request was successful
    response.raise_for_status()  # Raises an error for bad responses (4xx or 5xx)

    # Check if response content is not empty
    if response.content:
        # Print the prediction
        prediction = response.json()
        print("Prediction:", prediction)
    else:
        print("Error: Empty response received from the API.")

except requests.exceptions.RequestException as e:
    print(f"Error during API request: {e}")
    sys.exit(1)
except ValueError as ve:
    print(f"Error parsing JSON response: {ve}")
    sys.exit(1)
```
